```
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(0,2*np.pi,0.1)

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '-')


plt.title("Sample Visualization")
plt.grid()
plt.show()
```
